```python
import matplotlib.pyplot as plt
# 13 TeV points
x1 = [0,1,2,3,4]
y1 = [800.249 ,687.967, 678.098, 646.489, 559.546]
y1error=[135.035, 213.686, 197.301, 200.888, 153.813]

# plotting the line 1 points 
plt.plot(x1, y1, label = "13 TeV")

# 13.6 TeV points
x2 = [0,1,2,3,4]
y2 = [950.408, 845.975, 660.182, 778.598, 502.913]
y2error = [148.71, 228.663, 447.529, 205.135,	165.849]

# plotting the line 2 points 
plt.plot(x2, y2, label = "13.6 TeV")

# 14 TeV points
x3 = [0,1,2,3,4]
y3 = [964.722, 832.107, 691.498, 606.89, 503.918]
y3error = [297.965, 176.272, 323.205, 183.476, 155.725]

# plotting the line 3 points 
plt.plot(x3, y3, label = "14 TeV")

plt.xlabel('Jet multiplicity (NJETs)', weight="bold")
# Set the y axis label of the current axis.
plt.ylabel('$\sigma_{tt}$ (pb)')
# Set a title of the current axes.
plt.title(r'NNLO $t\bar{t}$ Cross sections at LHC ($\sqrt{s}$=13,13.6 and 14 TeV)',weight="bold")
# Error bars
plt.errorbar(x1, y1, yerr=y1error, fmt='o', ecolor='black', capsize=4)
plt.errorbar(x2, y2, yerr=y2error, fmt='o', ecolor='black', capsize=4)
plt.errorbar(x3, y3, yerr=y3error, fmt='o', ecolor='black', capsize=4)

plt.figure(figsize=(50,50))

# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```
